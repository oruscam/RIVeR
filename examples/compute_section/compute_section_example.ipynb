## Compute section example

```python
import json
from pathlib import Path

from matplotlib import pyplot as plt

import river.core.compute_section as cs
```

```python
path_x_sections = Path('xsections.json')
path_results_piv = Path('piv_results.json')
path_transformation_matrix = Path('uav_transformation_matrix.json')

xsections = json.loads(path_x_sections.read_text())
piv_results = json.loads(path_results_piv.read_text())
transformation_matrix = json.loads(path_transformation_matrix.read_text())

step = 1
fps = 30
id_section = 0
```

```python
xsections = cs.update_current_x_section(xsections, piv_results, transformation_matrix, step, fps, id_section, interpolate = True)
```

```python
image = plt.imread('0000000003.jpg')
plt.imshow(image)

plt.plot(
    xsections['CS_default_1']['x'],
    xsections['CS_default_1']['y']
)

plt.quiver(
    xsections['CS_default_1']['x'],
    xsections['CS_default_1']['y'],
    xsections['CS_default_1']['displacement_x_streamwise'],
    xsections['CS_default_1']['displacement_y_streamwise'],
    color='red'
)
```
